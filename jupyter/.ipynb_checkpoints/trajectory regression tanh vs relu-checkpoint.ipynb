```python
import numpy as np
from matplotlib import pyplot as plt
import os
import pickle
import json
from copy import deepcopy
import pandas as pd
import sys
sys.path.insert(0, '../')
sys.path.insert(0, '../../')
from scipy.sparse.linalg import lsqr
from style import style_setup

os.system('python ../style/style_setup.py')
```

```python
common_path = '/Users/tolmach/Documents/GitHub/latent_circuit_inference/data/inferred_LCs'
# get trajectories from ReLU nets
relu_LC_folder = os.path.join(common_path, "CDDMrelu",
                              '0.007339_CDDM;relu;N=93;lmbdo=0.3;orth_inp_only=True;lmbdr=0.5;lr=0.005;maxiter=1000', 
                              '6_nodes',
                              '0.8479339516810706_0.8997970755148001_LC_6units_CDDMrelu')
relu_LC_trajectories = pickle.load(open(os.path.join(relu_LC_folder, "0.007339_LCtrajdata_CDDM.pkl"), 'rb+'))["trajectories"]

tanh_LC_folder = os.path.join(common_path, "CDDMtanh",
                              '0.0111542_CDDM;tanh;N=100;lmbdo=0.3;lmbdr=0.5;lr=0.005;maxiter=1500', 
                              '6_nodes',
                              '0.8881552431856446_0.9446767795389052_LC_6units_CDDMtanh')
tanh_LC_trajectories = pickle.load(open(os.path.join(tanh_LC_folder, "0.0111542_LCtrajdata_CDDM.pkl"), 'rb+'))["trajectories"]
```

```python
T = relu_LC_trajectories.shape[1]
K = relu_LC_trajectories.shape[2]
relu_mat = relu_LC_trajectories.reshape(-1, T*K)
tanh_mat = tanh_LC_trajectories.reshape(-1, T*K)
```

```python
A = np.zeros((6,6))

for i in range(6):
    A[:, i] = lsqr(relu_mat.T, tanh_mat[i, :], damp = 0)[0]
    
B = np.zeros((6,6))

for i in range(6):
    B[:, i] = lsqr(tanh_mat.T, relu_mat[i, :], damp = 0)[0]
```

```python
scoreA = np.mean((A.T @ relu_mat  - tanh_mat)**2) / np.mean((tanh_mat)**2)

scoreB = np.mean((B.T @ tanh_mat  - relu_mat)**2) / np.mean((relu_mat)**2)
```

```python
print(scoreA)
print(scoreB)
```

```python
tanh_LC_trajectories_predicted = np.einsum("ij,jtk->itk", A.T, relu_LC_trajectories)
relu_LC_trajectories_predicted = np.einsum("ij,jtk->itk", B.T, tanh_LC_trajectories)
```

```python
k = 10
i = 4
fig, axs = plt.subplots(6,1 , figsize = (5, 8))
for i in range(6):
    axs[i].plot(tanh_LC_trajectories_predicted[i, :, k], color='r')
    axs[i].plot(tanh_LC_trajectories[i, :, k], color='b')
plt.show()
```

```python
k = 10
i = 4
fig, axs = plt.subplots(6,1 , figsize = (5, 8))
for i in range(6):
    axs[i].plot(relu_LC_trajectories_predicted[i, :, k], color='r')
    axs[i].plot(relu_LC_trajectories[i, :, k], color='b')
plt.show()
```
