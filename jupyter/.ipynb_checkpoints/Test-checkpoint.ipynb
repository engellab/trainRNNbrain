```python
import typing as T
import os
import matplotlib.colors as mcolors
import matplotlib.font_manager as mfonts


def setup_fonts(fontpaths: list[os.PathLike], verbose: bool = False, **kwargs):
  for font in sorted(mfonts.findSystemFonts(fontpaths=fontpaths, **kwargs)):
    mfonts.fontManager.addfont(font)
    if verbose: print(f'font:\t{os.path.basename(font)}')

def setup_colors(themepaths: list[os.PathLike], verbose: bool = False):
  for themepath in themepaths:
    with open(themepath, 'rt') as f:
      lines = f.readlines()
      for line in lines:
        line = line.strip()
        if line.startswith('#'): continue
        names, value = line.split(':')
        value = value.strip()
        for name in names.split(','):
          name = name.strip()
          mcolors._colors_full_map[name] = value  # type: ignore
          if verbose: print(f'color:\t{name:16s} {value:}')
```

```python
# Setup plotting.
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
# import utils.plot as utp
%config InlineBackend.figure_format = 'retina'
setup_fonts(['assets/fonts'], verbose=True)
setup_colors(['assets/styles/scientific.txt'], verbose=True)
plt.style.use(['assets/styles/scientific.mplstyle', { 'figure.dpi': 144 }])
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
mm = 1/25.1
time = 1000
a = np.log(40 + np.arange(time) + 20*np.random.randn(time))

fig, ax = plt.subplots(1,1, figsize = (3, 1.5))
ax.plot(a, color = 'r')
ax.set_xlabel("Numbers")
ax.set_ylabel("Other Numbers")

plt.show()
```

```python
A = 3 * np.random.randn(100, 2) + 5.5 * np.random.randn(2)
B = 1.5 * np.random.randn(100, 2) + 5.5 * np.random.randn(2)
C = 0.5 * np.random.randn(100, 2) + 5.5 * np.random.randn(2)
```

```python
fig, ax = plt.subplots(1,1, figsize = (40 * mm, 40 * mm))

ax.scatter(A[:, 0], A[:, 1], color = 'yellow', edgecolor = 'k')
ax.scatter(B[:, 0], B[:, 1], color = 'cyan', edgecolor = 'k')
ax.scatter(C[:, 0], C[:, 1], color = 'magenta', edgecolor = 'k')
ax.set_xlabel("Numbers")
ax.set_ylabel("Other Numbers")

plt.show()
```
