```python
import numpy as np
from matplotlib import pyplot as plt
import os
import sys
import json
import pickle
from src.utils import jsonify
import pandas as pd
import datetime
```

```python
task_name = "CDDM"
folder_path = os.path.join("../", "data", "trained_RNNs", task_name)
folders = os.listdir(folder_path)
new_folder_names = []
folders_filtered = []
for folder in folders:
    if folder == '.DS_Store' or folder == 'exemplar_RNN':
        pass
    else:
        day = float(datetime.datetime.fromtimestamp(os.path.getmtime(os.path.join(folder_path, folder))).strftime('%d'))
        month = float(datetime.datetime.fromtimestamp(os.path.getmtime(os.path.join(folder_path, folder))).strftime('%m'))
        year = float(datetime.datetime.fromtimestamp(os.path.getmtime(os.path.join(folder_path, folder))).strftime('%Y'))
        if  (day > 19) and (month > 5):
            folders_filtered.append(folder)
```

```python
len(folders_filtered)
```

```python
scores = []
for folder in folders_filtered:
        files = os.listdir(os.path.join(folder_path, folder))
        config_file = None
        for file in files:
            if "config" in file:
                config_file = file
        config_data = json.load(open(os.path.join(folder_path, folder, config_file), "rb+"))
        score = np.round(float(config_file.split("_")[0]), 7)
        activation = config_data["activation"]
        N = config_data["N"]
        print(N)
        lmbdo = config_data["lambda_orth"]
        lmbdr = config_data["lambda_r"]
        lr=config_data["lr"]
        maxiter=config_data["max_iter"]
        extra_info = f"{task_name};{activation};N={N};lmbdo={lmbdo};lmbdr={lmbdr};lr={lr};maxiter={maxiter}"
        name = f"{score}_{extra_info}"
        new_folder_names.append(name)
        scores.append(score)
```

```python
df = pd.DataFrame({"old_name" : folders_filtered, "new_name": new_folder_names, "scores" : scores})
```

```python
pd.set_option('display.max_rows', None)
df.sort_values("scores")
```

```python
df.drop_duplicates(subset='new_name', inplace=True)
```

```python
folders_old = df["old_name"].to_list()
folders_new = df["new_name"].to_list()
```
