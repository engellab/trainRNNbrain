what would be the first principal component of the bunch of point where only one neuron fires at a time?

```python
import numpy as np
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
```

```python
n = 30
r = 0.01
e = 1e-1
A = np.zeros((n, n))
for i in range(n):
    A[i, (i - 1)%n] = r * (1 + e * np.random.randn()) #1e-6
    A[i, i] = 1
    A[i, (i + 1)%n] = r * (1 + e * np.random.randn())#1e-6
    
B = np.zeros((n, n))
for i in range(n):
    B[i, (i - 1)%n] = r#1e-6
    B[i, i] = 1
    B[i, (i + 1)%n] = r#1e-6
B[0, -1] = 0
B[-1, 0] = 0
B[0, 0] = 1 + r
B[-1, -1] = 1 + r
```

```python
fig = plt.figure(figsize = (6,6))
plt.imshow(A)
plt.show()
```

```python
pA = PCA(n_components=6)
pA.fit(A)
projected_A = pA.components_ @ A

pB = PCA(n_components=6)
pB.fit(B)
projected_B = pB.components_ @ B
```

```python
fig = plt.figure(figsize = (6,6))
plt.scatter(projected_A[0, :], projected_A[1, :], color='r', edgecolor='k', s = 15)
plt.show()
```
